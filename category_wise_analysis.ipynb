```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('Datasets/Item-wise sales.xlsx')
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
# Group by Category and calculate total quantities for each quarter
category_trends = df.groupby("Category")[[
    "Total Quantity(Oct-Dec23)",
    "Total Quantity(Jan-Mar24)",
    "Total Quantity (Apr-Jun24)",
    "Total Quantity (Jul-Sep24)"
]].sum()

# Rename columns for better readability
category_trends.columns = ["Oct-Dec23", "Jan-Mar24", "Apr-Jun24", "Jul-Sep24"]

# Transpose the data for plotting
category_trends = category_trends.T

# Plotting the trend
plt.figure(figsize=(10, 6))
for category in category_trends.columns:
    plt.plot(category_trends.index, category_trends[category], marker='o', label=category)

# Customize the plot
plt.title("Category-Wise Trend Analysis Over 4 Quarters", fontsize=14)
plt.xlabel("Quarters", fontsize=12)
plt.ylabel("Total Quantity", fontsize=12)
plt.grid(True)
plt.legend(title="Category", fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
