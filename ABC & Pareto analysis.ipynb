```python
#Remove NaN values
data_cleaned = data.dropna()
```

```python
data_cleaned.head()
```

```python
data_sorted1 = data_cleaned.sort_values(by="Revenue", ascending=False).reset_index(drop=True)
```

```python
data_sorted1.head()
```

```python
# Step: Calculate Cumulative Revenue and Percentage
data_sorted1["Cumulative Revenue"] = data_sorted1["Revenue"].cumsum()
data_sorted1["Cumulative Percentage"] = (data_sorted1["Cumulative Revenue"] / data_sorted1["Revenue"].sum()) * 100 
```

```python
data_sorted1.head()
```

```python
# Step: Classify items into A, B, C categories
def classify(row):
    if row["Cumulative Percentage"] <= 80.000000:
        return "A"
    elif row["Cumulative Percentage"] <= 95.000000:
        return "B"
    else:
        return "C"

# Apply the classification
data_sorted1["ABC Category"] = data_sorted1.apply(classify, axis=1) #axis=1 means: Apply the function row-wise.
```

```python
data_sorted1.head()
```

```python
# Step 1: Analyze counts and percentages
category_counts = data_sorted1["ABC Category"].value_counts()
category_percentages = (category_counts / len(data_sorted1)) * 100

# Print counts and percentages
print("Counts per category:")
print(category_counts)
print("\nPercentages per category:")
print(category_percentages)

# Step 2: Visualize the distribution with a pie chart
plt.figure(figsize=(8, 6))
category_counts.plot.pie(autopct='%1.1f%%', labels=category_counts.index, colors=["#66c2a5", "#fc8d62", "#8da0cb"])
plt.title("ABC Category Distribution")
plt.ylabel("")  # Hide the y-label for a clean look
plt.show()
```

```python
# Step: Analyze revenue contribution by category
revenue_by_category = data_sorted1.groupby("ABC Category")["Revenue"].sum()
revenue_percentages = (revenue_by_category / revenue_by_category.sum()) * 100

# Print the results
print("Revenue contribution by category:")
print(revenue_by_category)
print("\nRevenue percentages:")
print(revenue_percentages)

# Visualize the revenue contribution with a pie chart
plt.figure(figsize=(8, 6))
revenue_by_category.plot.pie(autopct='%1.1f%%', labels=revenue_by_category.index, colors=["#8da0cb", "#fc8d62", "#66c2a5"])
plt.title("Revenue Contribution by ABC Category")
plt.ylabel("")  # Hide the y-label for a clean look
plt.show()
```
